```python
from matplotlib import pyplot as plt

# Grouping data
grouped_data = df1.groupby('Nation')['Player'].count()

# Plotting horizontal bar chart
plt.figure(figsize=(10, 6))  # Set figure size
plt.barh(grouped_data.index, grouped_data.values, color='skyblue', edgecolor='black')

# Adding labels and title
plt.title("Number of Players by Nation", fontsize=12)
plt.xlabel("Number of Players", fontsize=1)
plt.ylabel("Nation", fontsize=1)
# plt.grid(axis='x', linestyle='--', alpha=0.7)  # Optional: Add grid lines for better readability

# Show plot
plt.show()
```
